```python
svc_accuracy = 0.8262004175365344

model_names = ['Logistic Regression','Neural Network \n(31602 parameters)','Support Vector Machine','Random Forest', 'Ada Boost', 'Gradient Boosting', 'Bagging', 'Voting classifier \n(All other ensembles)']
accuracies = [ 0.6017745302713987,0.7406054279749478, 0.8262004175365344, 0.839405419253558, 0.836587701750274, 0.840137361888617, 0.8374216301367623, 0.8384661800857967]

model_accuracy_dict = dict(zip(model_names, accuracies))
print(model_accuracy_dict)
```

```python
import matplotlib.pyplot as plt

model_names = list(model_accuracy_dict.keys())
accuracies = list(model_accuracy_dict.values())

plt.figure(figsize=(10, 6))  # Adjust the width as per your requirement
plt.bar(model_names, accuracies)
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.xticks(rotation=45)

# Add text annotations to each bar
for i, v in enumerate(accuracies):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom')

plt.show()
```

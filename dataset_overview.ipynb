```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'data/loan_data.csv'
column_names = ['credit.policy', 'purpose', 'int.rate', 'installment', 'log.annual.inc', 'dti', 'fico', 'days.with.cr.line', 'revol.bal', 'revol.util', 'inq.last.6mths', 'delinq.2yrs', 'pub.rec', 'Class']
data : pd.DataFrame = pd.read_csv(url, names =column_names, na_values="?")
data = data.dropna()
```

```python
numerical_columns = ['int.rate', 'installment', 'log.annual.inc', 'dti', 'fico', 'days.with.cr.line', 'revol.bal', 'revol.util', 'inq.last.6mths', 'pub.rec']

fig, axes = plt.subplots(nrows=1, ncols=len(numerical_columns), figsize=(15, 5))

for i, column in enumerate(numerical_columns):
    axes[i].boxplot(data[column])
    axes[i].set_title(column)
    axes[i].set_xlabel('Values')
    axes[i].set_ylabel(column)

plt.tight_layout()
plt.show()
```
